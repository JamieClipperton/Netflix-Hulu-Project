## Extract

```python
#Import dependencies
import os
import pandas as pd
from sqlalchemy import create_engine
```

```python
#File Names
TitlesFile = os.path.join("CSVfiles","titles.csv")
CreditsFile = os.path.join("CSVfiles","credits.csv")
```

```python
titles_df = pd.read_csv(TitlesFile)

credits_df = pd.read_csv(CreditsFile)
```

## Transform

```python
# Clean titles_df
titles_df.columns
```

```python
# We want to drop show_id, cast, country, date_added
del titles_df['age_certification']
del titles_df['runtime']
del titles_df['production_countries']
del titles_df['imdb_id']
del titles_df['imdb_votes']
del titles_df['tmdb_popularity']
del titles_df['tmdb_score']
```

```python
# Rename columns 
titles_df = titles_df.rename(columns={"description":"overview"})
```

```python
# Clean credits_df
credits_df.columns
```

```python
# We want to drop person_id, role
del credits_df['person_id']
del credits_df['role']
```

```python
# Rename columns
credits_df = credits_df.rename(columns={"name":"name of actor/actress"})
```

```python
# Merge dfs
netflix_df = pd.concat([titles_df, credits_df])
netflix_df
```
