```python
import pandas as pd
import datetime as dt
```

```python
#load files
titles = "CSVfiles/titles.csv"
credits = "CSVfiles/credits.csv"
titles_file = pd.read_csv(titles)
credits_file = pd.read_csv(credits)
#merge files
netflix_data_df = pd.merge(titles_file, credits_file, on="id")
netflix_data_df.head()
```

```python
#sort data so that only tv shows are in a Dataframe
tv_shows_df=netflix_data_df.loc[netflix_data_df["type"] == "SHOW", :]
tv_shows_df.head()
```

```python
#create data frame with important information
shows_df = tv_shows_df[['title','description','release_year','age_certification','runtime','seasons','name','character','role']].copy()
clean_shows_df = shows_df.drop_duplicates(["title"])
clean_shows_df
```

```python
#dropped NaN results
clean_shows_df= clean_shows_df.dropna(how='any')
clean_shows_df
```
